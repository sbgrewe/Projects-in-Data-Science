```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
penguins = sns.load_dataset("penguins")
```

```python
# Apply the default theme
sns.set_theme()

# Load an example dataset
tips = sns.load_dataset("tips")

# Create a visualization
sns.relplot(
    data=tips,
    x="total_bill", y="tip", col="time",
    hue="smoker", style="smoker", size="size",
)
```

Here is an example of how to overwrite fill color which is dependent on a categorical variable:

```python
muted    = ["#4878CF", "#6ACC65", "#D65F5F", "#B47CC7", "#C4AD66", "#77BEDB"]
newPal   = {'Chinstrap': muted[4], 'Adelie': muted[2], 'Gentoo': muted[0]}
sns.displot(penguins, x="flipper_length_mm", hue="species", binwidth = 3, palette=newPal, multiple="stack")
```
